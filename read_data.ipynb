```python
import segyio
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# 2D marmousi model 
nptz = 751
nptx = 2301
with segyio.open('marmousi/velocity_rev1.segy', strict=False) as f:
    trace = f.trace
    n_traces = len(trace)
    n_samples = len(trace[0])
    assert nptz <= n_traces and nptx <= n_samples
    z = np.linspace(0, n_traces-1, nptz, endpoint=True).astype(np.int)
    x = np.linspace(0, n_samples-1, nptx, endpoint=True).astype(np.int)
    v = tuple()
    for j in z:
        vj = trace[j][x]
        v = v + (vj,)
    v = np.vstack(v)
```

```python
plt.imshow(v)
plt.title('velocity in marmousi model')
```

```python
# 2D 2004 BP model
nptz = 1911
nptx = 5395
with segyio.open('bp2004/vel_z6.25m_x12.5m_exact.segy', strict=False) as f:
    trace = f.trace
    n_traces = len(trace)
    n_samples = len(trace[0])
    z = np.linspace(0, n_samples-1, nptz, endpoint=True).astype(np.int)
    x = np.linspace(0, n_traces-1, nptx, endpoint=True).astype(np.int)
    assert nptz <= n_samples and nptx <= n_traces
    v = tuple()
    for i in x:
        vi = trace[i][z]
        v = v + (vi,)
    v = np.column_stack(v)
```

```python
plt.imshow(v)
plt.title('velocity in 2004 BP model')
```
